##### Remarque : On observe différents montant et certains , les prix (montant) sont n'a pas tous de même unité , exemple 80 correspond normalement à 80000. On va effectuer des transformations afin de les rendre tous à la même echelle.  

```python
# Opération sur la variable montant
def Transform(a):
    if a==35:
        return 35000
    elif a==200:
        return 200000
    elif a==500:
        return 500000
    elif a in range(2,6):
        return a*10000
    # elif a==5:
    #     return 50000
    elif a==80:
        return 80000
    elif a==450:
        return 450000
    elif a==50:
        return 50000
    else:
        return a
data['montant']=data['montant'].apply(Transform)
data['montant'].unique()
```
